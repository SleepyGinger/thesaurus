```python
import requests
from bs4 import BeautifulSoup
import time
from IPython.display import clear_output

def syn_word(word):
    response = requests.get(f"https://www.thesaurus.com/browse/{word}")
    soup = BeautifulSoup(response.content, 'html.parser')

    # select only first '<section class="synonyms-container....'
    synonyms = soup.select('#initial-load-content > main > section > section > div:nth-child(2) > ul > li')

    syn_list = []

    for synonym in synonyms:
        syn_list.append(synonym.text)
        #print synonym.get('href')
    
    return {word:syn_list}

def loop_it(words_and_syn):
    for word in my_word_list[6:]:
        clear_output(wait=True)
        new = syn_word(word)
        words_and_syn = {**words_and_syn,**new}
        print(words_and_syn.keys())
        print(words_and_syn[word])
        time.sleep(10)
    return words_and_syn
```

```python
from nltk.corpus import wordnet
synonyms = []

for syn in wordnet.synsets("learning"):
    for l in syn.lemmas():
        synonyms.append(l.name())

print(set(synonyms))
```

```python
def check_word(word):
    url=f'https://words.bighugelabs.com/api/2/8fd84ed65f4661af1c474da7f6d526cc/{word}/json'

    r = requests.get(url)
    result = r.json()

    syn=[]

    for i in result:
        for x in result[i]:
            for y in result[i][x]:
                syn.append(y)
                
    return syn

check_word('learning')
```
